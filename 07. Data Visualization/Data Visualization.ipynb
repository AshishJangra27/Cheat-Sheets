# Data Visualization with Python Cheat Sheet

This cheat sheet covers the essential and advanced features of data visualization using Python libraries such as Matplotlib, Seaborn, and Plotly. It includes examples and code snippets to help you get started with various types of plots and customization options.

## 1. Introduction to Data Visualization
### 1.1 What is Data Visualization?
Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

### 1.2 Importance of Data Visualization
Data visualization is important because it helps to:
- Understand data quickly
- Identify patterns, trends, and outliers
- Communicate data insights effectively
- Support decision making


### 1.3 Libraries for Data Visualization in Python
Python offers several libraries for data visualization, including:
- **Matplotlib**: A versatile and powerful library for creating static, animated, and interactive visualizations.
- **Seaborn**: A statistical data visualization library based on Matplotlib, which provides a high-level interface for drawing attractive and informative statistical graphics.
- **Plotly**: A graphing library that makes interactive, publication-quality graphs online. It includes APIs for Python, R, and MATLAB.

### 1.4 Overview of Data Visualization Process
The data visualization process typically involves the following steps:
1. Data Collection: Gather data from various sources.
2. Data Cleaning: Clean and preprocess the data to handle missing values, outliers, and errors.
3. Data Exploration: Explore the data to understand its structure and characteristics.
4. Data Visualization: Create visual representations of the data to identify patterns, trends, and insights.
5. Interpretation: Interpret the visualizations to derive meaningful insights and support decision making.


## 2. Matplotlib
Matplotlib is a versatile and powerful library for creating static, animated, and interactive visualizations in Python. It is highly customizable and supports various types of plots.

### 2.1 Introduction to Matplotlib
To get started with Matplotlib, you need to install it using pip:
```
pip install matplotlib
```
Once installed, you can import it in your script:


````python
import matplotlib.pyplot as plt
# Now you can use Matplotlib functions with the plt prefix
````

### 2.2 Basic Plots
Matplotlib supports various basic plots such as line plots, scatter plots, bar plots, histograms, and pie charts.

#### 2.2.1 Line Plot

````python
# Creating a line plot
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
````

#### 2.2.2 Scatter Plot

````python
# Creating a scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

#### 2.2.3 Bar Plot

````python
# Creating a bar plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

#### 2.2.4 Histogram

````python
# Creating a histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

#### 2.2.5 Pie Chart

````python
# Creating a pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
````

#### 2.3.1 Subplots

````python
# Creating subplots
fig, axs = plt.subplots(2, 2)
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Plot 1')
axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Plot 2')
axs[1, 0].bar(categories, values)
axs[1, 0].set_title('Plot 3')
axs[1, 1].hist(data, bins=30)
axs[1, 1].set_title('Plot 4')
plt.tight_layout()
plt.show()
````

#### 2.3.2 Error Bars

````python
# Creating a plot with error bars
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.1
plt.errorbar(x, y, fmt='-')
plt.title('Error Bars')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
````

#### 2.3.3 3D Plots

````python
# Creating a 3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Plot')
plt.show()
````

#### 2.3.4 Stacked Bar Chart

````python
# Creating a stacked bar chart
labels = ['A', 'B', 'C', 'D']
men_means = [20, 35, 30, 35]
women_means = [25, 32, 34, 20]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x, men_means, width, label='Men')
ax.bar(x, women_means, width, bottom=men_means, label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
````

### 2.4 Customization
Matplotlib allows you to customize your plots in various ways, including adding titles and labels, legends, annotations, and changing colors and styles.

#### 2.4.1 Titles and Labels

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

#### 2.4.2 Legends

````python
# Adding legends to a plot
plt.plot(x, y, label='Sine Wave')
plt.plot(x, np.cos(x), label='Cosine Wave')
plt.title('Plot with Legends')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````

#### 2.4.3 Annotations

````python
# Adding annotations to a plot
plt.plot(x, y)
plt.title('Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.annotate('Max Point', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
````

#### 2.4.4 Colors and Styles

````python
# Customizing colors and styles
plt.plot(x, y, 'r--', label='Sine Wave')  # Red dashed line
plt.plot(x, np.cos(x), 'b:', label='Cosine Wave')  # Blue dotted line
plt.title('Customized Colors and Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````

#### 2.4.5 Themes

````python
# Applying a theme to your plots
plt.style.use('ggplot')
plt.plot(x, y)
plt.title('Plot with ggplot Theme')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

### 3.3 Advanced Plots
Seaborn also supports various advanced plots such as box plots, violin plots, heatmaps, pair plots, and facet grids.

#### 3.3.1 Box Plot

````python
categories = ['Group A', 'Group B', 'Group C']
values = [
    [10, 12, 15, 18, 20, 22, 25],  # Values for Group A
    [15, 18, 20, 22, 25, 28, 30, 32],  # Values for Group B
    [20, 25, 28, 30, 35, 40, 42]  # Values for Group C
]

# Flatten the values list
flattened_values = np.concatenate(values)

# Create a corresponding list of categories for each value
flattened_categories = np.repeat(categories, [len(v) for v in values])

sns.boxplot(x=flattened_categories, y=flattened_values)
plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

#### 3.3.2 Violin Plot

````python
# Creating a violin plot
sns.violinplot(x=flattened_categories, y=flattened_values)
plt.title('Violin Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

#### 3.3.3 Heatmap

````python
# Creating a heatmap
data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, fmt='0.1f')
plt.title('Heatmap')
plt.show()
````

#### 3.3.4 Pair Plot

````python
# Creating a pair plot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.title('Pair Plot')
plt.show()
````

#### 3.3.5 Facet Grid

````python
# Creating a facet grid
g = sns.FacetGrid(iris, col='species')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
plt.show()
````

#### 3.3.6 Categorical Plots

````python
# Creating categorical plots
tips = sns.load_dataset('tips')
sns.catplot(x='day', y='total_bill', hue='smoker', kind='swarm', data=tips)
plt.title('Categorical Plot')
plt.show()
````

````python
# Creating a line plot
fig = px.line(x=x, y=y, title='Line Plot')
fig.show()
````

#### 4.2.2 Scatter Plot

````python
# Creating a scatter plot
fig = px.scatter(x=x, y=y, title='Scatter Plot')
fig.show()
````

#### 4.2.3 Bar Plot

````python
# Creating a bar plot
categories = ['Group A', 'Group B', 'Group C']
values = [15, 28, 35]

fig = px.bar(x=categories, y=values, title='Bar Plot')
fig.show()
````

#### 4.2.4 Histogram

````python
# Creating a histogram

data = np.random.randn(1000)

fig = px.histogram(x=data, title='Histogram')
fig.show()
````

### 4.3 Advanced Plots
Plotly also supports various advanced plots such as 3D plots, box plots, violin plots, and heatmaps.

#### 4.3.1 3D Plot

````python
# Creating a 3D plot
fig = go.Figure(data=[go.Surface(z=np.sin(np.sqrt(X**2 + Y**2)))])
fig.update_layout(title='3D Plot', scene=dict(zaxis_title='Z'))
fig.show()
````

#### 4.3.2 Box Plot

````python
categories = ['Group A', 'Group B', 'Group C']
values = [
    [10, 12, 15, 18, 20, 22, 25],  # Values for Group A
    [15, 18, 20, 22, 25, 28, 30, 32],  # Values for Group B
    [20, 25, 28, 30, 35, 40, 42]  # Values for Group C
]

# Flatten the values list
flattened_values = np.concatenate(values)

# Create a corresponding list of categories for each value
flattened_categories = np.repeat(categories, [len(v) for v in values])

fig = px.box(x=flattened_categories, y=flattened_values, title='Box Plot')
fig.show()
````

#### 4.3.3 Violin Plot

````python
# Creating a violin plot
fig = px.violin(x=flattened_categories, y=flattened_values, title='Violin Plot')
fig.show()
````

#### 4.3.4 Heatmap

````python
# Creating a heatmap
fig = px.imshow(np.random.rand(10, 12), title='Heatmap')
fig.show()
````

### 4.4 Customization
Plotly allows you to customize your plots in various ways, including adding titles and labels, legends, annotations, and changing colors and styles.

#### 4.4.1 Titles and Labels

````python
# Adding titles and labels to a plot
fig = px.line(x=x, y=y, title='Customized Plot')
fig.update_layout(title='Customized Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
````

#### 4.4.2 Legends

````python
# Adding legends to a plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Sine Wave'))
fig.add_trace(go.Scatter(x=x, y=np.cos(x), mode='lines', name='Cosine Wave'))
fig.update_layout(title='Plot with Legends', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
````

#### 4.4.3 Annotations

````python
# Adding annotations to a plot
fig = px.line(x=x, y=y, title='Plot with Annotations')
fig.add_annotation(x=np.pi/2, y=1, text='Max Point', showarrow=True, arrowhead=1)
fig.update_layout(xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
````

#### 4.4.4 Colors and Styles

````python
# Customizing colors and styles
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Sine Wave', line=dict(color='red', dash='dash')))
fig.add_trace(go.Scatter(x=x, y=np.cos(x), mode='lines', name='Cosine Wave', line=dict(color='blue', dash='dot')))
fig.update_layout(title='Customized Colors and Styles', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
````

#### 4.4.5 Themes

````python
# Applying a theme to your plots
fig = px.line(x=x, y=y, title='Plot with Plotly Theme')
fig.update_layout(template='plotly_dark')
fig.update_layout(xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
````

## 5. Interactive Plots
Interactive plots allow users to explore data dynamically. This section covers creating interactive plots using Plotly.

### 5.1 Introduction to Interactive Plots
Interactive plots are particularly useful for exploring data and gaining insights. Plotly provides various tools to create interactive visualizations.

#### 5.1.1 Adding Interactivity

````python
# Adding interactivity to a plot
fig = px.scatter(x=x, y=y, title='Interactive Scatter Plot')
fig.update_traces(marker=dict(size=12, opacity=0.6), selector=dict(mode='markers'))
fig.show()
````

#### 5.1.2 Hover Information

````python
import pandas as pd

import numpy as np

# Create a DataFrame with at least 20 rows
df = pd.DataFrame({
    'x': np.arange(1, 21),  # Numbers from 1 to 20
    'y': np.random.randint(10, 50, 20),  # 20 random integers between 10 and 50
    'additional_info': [chr(i) for i in range(65, 85)]  # Letters from 'A' to 'T'
})

# Create a scatter plot with hover data
fig = px.scatter(df, x='x', y='y', hover_data=['additional_info'])
fig.show()
````

#### 5.1.3 Zoom and Pan

````python
# Enabling zoom and pan
fig = px.line(x=x, y=y, title='Line Plot with Zoom and Pan')
fig.update_layout(dragmode='pan', hovermode='closest')
fig.show()
````

### 5.2 Advanced Customization
Plotly provides advanced customization options to create highly interactive and visually appealing plots.

#### 5.2.1 Subplots

````python
# Creating subplots
from plotly.subplots import make_subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=('Plot 1', 'Plot 2'))
fig.add_trace(go.Scatter(x=x, y=y, name='Sine Wave'), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=np.cos(x), name='Cosine Wave'), row=1, col=2)
fig.update_layout(title_text='Subplots')
fig.show()
````

#### 5.2.2 Customizing Hover Information

````python
# Customizing hover information
fig = px.scatter(x=x, y=y, title='Customized Hover Info')
fig.update_traces(marker=dict(size=12), hoverinfo='x+y')
fig.show()
````

#### 5.2.3 Adding Buttons and Dropdowns

````python
# Adding buttons and dropdowns
fig = px.scatter(x=x, y=y, title='Plot with Buttons')
fig.update_layout(updatemenus=[dict(type='buttons', showactive=False, buttons=[dict(label='Play', method='animate', args=[None, {'frame': {'duration': 500, 'redraw': True}, 'fromcurrent': True}]), dict(label='Pause', method='animate', args=[[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate', 'transition': {'duration': 0}}])])])
fig.show()
````

#### 5.2.4 Adding Sliders

````python
# Adding sliders
fig = px.scatter(x=x, y=y, title='Plot with Sliders')
fig.update_layout(sliders=[dict(steps=[dict(method='update', args=[{'visible': [True, False]}, {'title': 'Sine Wave'}], label='Sine Wave'), dict(method='update', args=[{'visible': [False, True]}, {'title': 'Cosine Wave'}], label='Cosine Wave')])])
fig.show()
````

### 5.3 Exporting Interactive Plots
Plotly allows you to export interactive plots in various formats such as HTML and JSON.

#### 5.3.1 Saving as HTML

````python
# Saving a plot as an HTML file
fig = px.line(x=x, y=y, title='Interactive Plot to Save')
fig.write_html('plotly_interactive_plot.html')
````

#### 5.3.2 Saving as JSON

````python
# Saving a plot as a JSON file
fig = px.line(x=x, y=y, title='Interactive Plot to Save')
fig.write_json('plotly_interactive_plot.json')
````

````python
# Folium is a Python library used for visualizing geospatial data. It allows you to create interactive maps with Leaflet.js.
# To get started with Folium, you need to install it using pip:
# pip install folium

# Once installed, you can create maps with Folium as follows:
import folium

# Create a map centered around a specific location
m = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

# Add a marker to the map
folium.Marker([45.5236, -122.6750], popup='Portland, OR').add_to(m)

# Save the map as an HTML file
m.save('map.html')
````
